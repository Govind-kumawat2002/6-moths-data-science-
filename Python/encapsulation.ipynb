```python
class Collage:
    collage_library = "yes"  ### public variable 
    _account_information = "2345678" ### private variable 
    college_name = "CIT"### public variable 
     
cr =Collage()
cr.collage_library
cr.account_information
```

```python
class Collage:
    __collage_library = "yes"  ### protected  variable 
    _account_information = "2345678" ### private variable 
    college_name = "CIT"### public variable 
     
cr =Collage()
cr.collage_library
# cr.account_information
```
