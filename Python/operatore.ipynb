###  Operators in Python

* **Operators** are *special symbols* used to perform operations on variables and values.

        Python Operators Names
1. Arithmetic Operators
2. Comparison Operators
3. Assignment Operators
4. Logical Operators
5. Membership Operators
6. Identity Operators
7. Bitwise Operators

## Arithmetic Operators
| Operator | Name                | Example   | Output |
| -------- | ------------------- | --------- | ------ |
| `+`      | Addition            | `10 + 5`  | `15`   |
| `-`      | Subtraction         | `10 - 5`  | `5`    |
| `*`      | Multiplication      | `10 * 5`  | `50`   |
| `/`      | Division            | `10 / 5`  | `2.0`  |
| `%`      | Modulus (Remainder) | `10 % 3`  | `1`    |
| `**`     | Exponent (Power)    | `2 ** 3`  | `8`    |
| `//`     | Floor Division      | `10 // 3` | `3`    |


```python
sum1 = 100 + 50      # 150 (100 + 50)
sum2 = sum1 + 250    # 400 (150 + 250)
sum3 = sum2 + sum2   # 800 (400 + 400)
print(sum3)
```

```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
print("Floor Division:", a // b)
```

### Assignment Operators

| Operator | Name                    | Example   | Same As      |
| -------- | ----------------------- | --------- | ------------ |
| `=`      | Assign                  | `x = 5`   | `x = 5`      |
| `+=`     | Add and Assign          | `x += 3`  | `x = x + 3`  |
| `-=`     | Subtract and Assign     | `x -= 3`  | `x = x - 3`  |
| `*=`     | Multiply and Assign     | `x *= 3`  | `x = x * 3`  |
| `/=`     | Divide and Assign       | `x /= 3`  | `x = x / 3`  |
| `%=`     | Modulus and Assign      | `x %= 3`  | `x = x % 3`  |
| `**=`    | Exponent and Assign     | `x **= 3` | `x = x ** 3` |
| `//=`    | Floor Divide and Assign | `x //= 3` | `x = x // 3` |


```python
# Assignment Operator
x = 10
print("Assign =", x)

# Add and Assign
x += 5
print("After += :", x)

# Subtract and Assign
x -= 3
print("After -= :", x)

# Multiply and Assign
x *= 2
print("After *= :", x)

# Divide and Assign
x /= 4
print("After /= :", x)

# Modulus and Assign
x %= 3
print("After %= :", x)

# Exponent and Assign
x **= 2
print("After **= :", x)

# Floor Division and Assign
x //= 2
print("After //= :", x)
```


### Comparison Operators in Python
| Operator | Name                  | Example  | Output  |
| -------- | --------------------- | -------- | ------- |
| `==`     | Equal to              | `5 == 5` | `True`  |
| `!=`     | Not Equal to          | `5 != 3` | `True`  |
| `>`      | Greater Than          | `5 > 3`  | `True`  |
| `<`      | Less Than             | `5 < 3`  | `False` |
| `>=`     | Greater Than Equal To | `5 >= 5` | `True`  |
| `<=`     | Less Than Equal To    | `3 <= 5` | `True`  |


```python
a = 10
b = 5

print("Equal to :", a == b)
print("Not Equal to :", a != b)
print("Greater Than :", a > b)
print("Less Than :", a < b)
print("Greater Than Equal To :", a >= b)
print("Less Than Equal To :", a <= b)
```

### Logical Operators 
| Operator | Name        | Example          | Output  |
| -------- | ----------- | ---------------- | ------- |
| `and`    | Logical AND | `True and False` | `False` |
| `or`     | Logical OR  | `True or False`  | `True`  |
| `not`    | Logical NOT | `not True`       | `False` |



```python
a = True
b = False

# Logical AND
print("AND Operator :", a and b)

# Logical OR
print("OR Operator :", a or b)

# Logical NOT
print("NOT Operator :", not a)
```

### Identity Operators
| Operator | Name                                                   | Example      | Output       |
| -------- | ------------------------------------------------------ | ------------ | ------------ |
| `is`     | Returns True if both variables are the same object     | `a is b`     | `True/False` |
| `is not` | Returns True if both variables are not the same object | `a is not b` | `True/False` |



```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

# is Operator
print("a is b :", a is b)

# is not Operator
print("a is not c :", a is not c)
```

### Membership Operators
| Operator | Name                                             | Example              | Output |
| -------- | ------------------------------------------------ | -------------------- | ------ |
| `in`     | Returns True if value is present in sequence     | `'a' in "apple"`     | `True` |
| `not in` | Returns True if value is not present in sequence | `'z' not in "apple"` | `True` |


```python
text = "python"

# in Operator
print("'p' in text :", 'p' in text)

# not in Operator
print("'z' not in text :", 'z' not in text)
```

### Bitwise Operators
| Operator | Name        | Example    | Output   |     |
| -------- | ----------- | ---------- | -------- | --- |
| `&`      | Bitwise AND | `5 & 3`    | `1`      |     |
| `        | `           | Bitwise OR | `5 \| 3` | `7` |
| `^`      | Bitwise XOR | `5 ^ 3`    | `6`      |     |
| `~`      | Bitwise NOT | `~5`       | `-6`     |     |
| `<<`     | Left Shift  | `5 << 1`   | `10`     |     |
| `>>`     | Right Shift | `5 >> 1`   | `2`      |     |


```python
a = 5
b = 3

# Bitwise AND
print("AND (&) :", a & b)

# Bitwise OR
print("OR (|) :", a | b)

# Bitwise XOR
print("XOR (^) :", a ^ b)

# Bitwise NOT
print("NOT (~) :", ~a)

# Left Shift
print("Left Shift (<<) :", a << 1)

# Right Shift
print("Right Shift (>>) :", a >> 1)
```
