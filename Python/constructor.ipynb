```python
from abc import ABC, abstractmethod

class School(ABC):

    @abstractmethod
    def show(self):
        pass

    def display(self):
        print("I am not hidden")


class Student(School):

    def show(self):
        print("Abstract Method Complete")


hello = Student()

hello.show()
hello.display()
```
