#### Conditional Statement in Python
            Definition

* A Conditional Statement in Python is used to make decisions in a program.
It executes different blocks of code based on whether a condition is True or False.

         Python provides these conditional statements:

* if
* if else
* if elif else
* nested if

| Statement      | Use                                             |
| -------------- | ----------------------------------------------- |
| `if`           | Executes code only if condition is true         |
| `if else`      | Executes one block if true and another if false |
| `if elif else` | Checks multiple conditions                      |
| `nested if`    | Uses one `if` statement inside another `if`     |
