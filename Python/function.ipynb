```python
# even odd 

def even_odd(lst):
    print("This is my list :- ",lst )
    for x in lst:
        if x%2==0:
            print("even", x )
        else:
            print("odd", x)
even_odd(lst= lst )
```

```python
# Create a function to calculate the sum of all elements in a list.
def cal_list(lst):
    total = 0 
    for x in lst:
        total+=x
    return total
    # pass 
cal_list(lst= [1,2,2,5,45,5])
```
