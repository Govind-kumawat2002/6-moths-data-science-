```python
# Multilevel Inheritance Example (Company Related)

# Parent Class
class Employee:
    def company_info(self):
        print("Company Name: TechSoft Pvt Ltd")


# Child Class
class Developer(Employee):
    def developer_work(self):
        print("Developer writes and tests code")


# Grand Child Class
class TeamLead(Developer):
    def manage_team(self):
        print("Team Lead manages the development team")


# Object Creation
obj = TeamLead()

# Accessing all methods
obj.company_info()
obj.developer_work()
obj.manage_team()
```
