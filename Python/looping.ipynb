#### Loop in Python

A loop is used to repeat a block of code multiple times until a condition becomes false.

For example:

* Print numbers from 1 to 10
* Print student names
* Repeat tasks automatically


Types of Loop in Python

| Loop Type    | Definition                                   |
| ------------ | -------------------------------------------- |
| `for loop`   | Used when the number of repetitions is known |
| `while loop` | Used when the condition is important         |
